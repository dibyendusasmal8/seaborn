## Why Seaborn?

Matplotlib, Plotly, and Seaborn are all popular Python libraries for data visualization, and each has its own strengths and use cases. While Matplotlib and Plotly are general-purpose plotting libraries, Seaborn is specifically designed for statistical data visualization. Here are some reasons why you might choose Seaborn in addition to or instead of Matplotlib and Plotly in the context of data science:

* bold Statistical Visualization:
Seaborn provides a high-level interface for drawing attractive and informative statistical graphics. It comes with several built-in themes and color palettes tailored for statistical data visualization. This makes it easier to create aesthetically pleasing plots without having to customize every detail manually.

* Default Aesthetics:
Seaborn sets good default aesthetics for plots, which often look more visually appealing than the default plots created with Matplotlib. This can save time and effort, especially when you need to quickly explore and visualize data during the initial stages of analysis.

* Statistical Plots:
Seaborn comes with a set of functions for creating informative statistical plots, such as violin plots, box plots, and swarm plots. These are designed to provide insights into the distribution of data and relationships between variables.

* Integration with Pandas:
Seaborn is built on top of Matplotlib and is designed to work seamlessly with Pandas DataFrames. This makes it easy to use Seaborn for visualizing data directly from Pandas structures.

* Conciseness and Simplicity:
Seaborn often requires fewer lines of code to create complex visualizations compared to Matplotlib. This can be advantageous when you need to create quick, exploratory plots or when you want to produce visually appealing plots without extensive customization.

* Combination with Matplotlib:
While Seaborn is high-level and easy to use, it can also be combined with Matplotlib for more fine-grained control over plot customization. Seaborn functions often return Matplotlib axes objects, allowing you to further modify the plots if needed.

That said, the choice between Matplotlib, Plotly, and Seaborn often depends on the specific requirements of your project, personal preferences, and the type of visualizations you need. Some data scientists prefer to use a combination of these libraries depending on the task at hand.

### 1. Relational Plot
In Seaborn, a relational plot is a type of plot that visualizes the relationship between two numerical variables. The <b>sns.relplot</b> function is a versatile tool for creating such plots, and it can be used to create scatter plots and line plots with ease

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## scatter plot 

```python
# scatter plot -> axes level function 
sns.scatterplot(data = tips ,x = 'total_bill',y = 'tip')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data = tips,x='total_bill',y='tip',kind='scatter')
```

### Adding more feature using multiple columns information

```python
# scatter plot -> axes level function 
sns.scatterplot(data = tips ,x = 'total_bill',y = 'tip',hue='sex',style='time',size='size')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data = tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

## line plot

```python
# axes level funtion
sns.lineplot(data=temp_gap,x='year',y='lifeExp')
```

### kdeplot
Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
# hue & fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# figursize manipulat using height and aspect parameter
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height = 10,aspect=2)
```

```python
tips.sample(2)
```

### Rugplot
Plot marginal distributions by drawing ticks along the x and y axes.
This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
# rugplot
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

### Bivariate histogram
A bivariate histogram bins the data within rectangles that tile the plot 
and then shows the count of observations within each rectangle with the fill color

```python
#bivarite histogram
sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# using figure laves (displot)
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

### Bivariate Kdeplot
a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot
Seaborn provides several functions for creating matrix plots, which are useful for visualizing relationships between multiple variables.<br>

Two commonly used functions are:
   - Heatmap
   - Clustermap

### Heatmap

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

### annot

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
# annot
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True)
```

```python
# linewidth
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

```python
# cmap
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
