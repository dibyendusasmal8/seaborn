## 1.Categorical Scatter Plot
This type of plot is useful for visualizing the distribution of data points along a category.
###  Stripplot 
Certainly! The <b >sns.stripplot </b>function in Seaborn is used to create a strip plot, which is a scatter plot where one of the variables is categorical. It is particularly useful for visualizing the distribution of data points along a category.

```python
# axes lavel funtion
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# Stripplot
sns.stripplot(data=tips,x='day',y='total_bill',jitter = False)
```

```python
# figure lavel function
sns.catplot(data=tips,x='day',y='total_bill',kind = 'strip')
```

```python
# jitter
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2)
```

In the context of Seaborn and strip plots, "jitter" refers to a small amount of random noise that is added to the horizontal position of each data point. This is done to avoid overlap when there are multiple data points with the same categorical value. The jitter parameter in the sns.stripplot function controls this behavior.

```python
# hue 
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```


In Seaborn, the hue parameter is used to add another categorical variable to a plot, allowing you to distinguish and color different levels of that variable. It is commonly used in various Seaborn functions, including sns.stripplot. The hue parameter helps create visualizations that highlight the relationship between two categorical variables or add an extra layer of information to the plot.

### swarmplot
The sns.swarmplot function in Seaborn is used to create a swarm plot, which is a categorical scatter plot with non-overlapping points. It is particularly useful for visualizing the distribution of data points along one or more categorical variables. 

```python
# axes lavel
sns.swarmplot(data=tips, x='day',y='total_bill')
```

```python
# figur lavel
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm',hue ='sex')
```

```python
# hue
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex')
```

## Categorical Distribution Plots
### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.


```python
# axes lavel funtion 
sns.boxplot(data=tips,y='total_bill')
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
# figur lavel 
# catplot 
sns.catplot(data=tips,x='sex',y='total_bill',kind='box')
```

```python
# hue
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',hue='day')
```

### Violinplot = (Boxplot + KDEplot)
A violin plot is a statistical data visualization that combines aspects of a box plot and a kernel density plot. It is particularly useful for visualizing the distribution of a quantitative variable across different levels of one or more categorical variables. Seaborn's <b>sns.violinplot</b> function can be used to create violin plots.

```python
# violinplot
# axes lavel
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# figure lavel 
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue

sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
# hue

sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

## barplot
A bar plot (or bar chart) is a common way to visualize the distribution of a categorical variable or the relationship between two categorical variables. In Seaborn, you can use the <b>sns.barplot</b> function to create bar plots

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# set estimator 
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',estimator=np.median)
```

In Seaborn's sns.barplot, the estimator parameter allows you to specify a function that is used to aggregate the data within each category and compute a summary statistic. This can be particularly useful when you want to show a summary measure other than the default, which is the mean
You can use any function that reduces an array to a scalar value, such as NumPy functions (np.mean, np.median, np.sum, etc.), or custom functions that you define.

```python
# hue
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
# ci
# ci parameter used for remove erore bar 

sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min,ci=None)
```

# point plot

In Seaborn, the<b> sns.pointplot</b> function is used to create point plots, which are similar to line plots but display the estimate of a central tendency for a numeric variable within categories. Point plots are particularly useful for showing how the central tendency changes across different levels of a categorical variable.

```python
sns.pointplot(data=tips, x='sex', y='total_bill')
```

```python
# after remove erore bar 
sns.pointplot(data=tips, x='sex', y='total_bill',ci=None)
```

 it's represent difference between two variables mean

```python
#hue
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',ci=None)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

# countplot
A countplot in Seaborn is a specialized bar plot that is designed for counting the occurrences of observations in each category of a categorical variable. It's a convenient way to visualize the distribution of a categorical variable

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',kind='box',col='smoker')
```

```python
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# figur level
# hue is working 
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

### residplot()
The <b>sns.residplot</b> function in Seaborn is used to create a residual plot, which is a scatter plot that shows the residuals on the vertical axis and the independent variable on the horizontal axis. Residuals represent the difference between the observed and predicted values in a regression analysis.

```python
sns.residplot(data=tips,x='total_bill',y='tip')
```

### A second way to plot Facet plots -> FacetGrid
In Seaborn, figure-level functions are those that create an entire figure or a complete set of subplots. These functions are typically more high-level and handle the creation and organization of the entire figure, including multiple subplots, rather than individual axes or plots. Some of the figure-level functions in Seaborn include <b>sns.relplot</b>,<b>sns.catplot</b>, <b>sns.lmplot</b>, and <b>sns.pairplot</b>

```python
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
